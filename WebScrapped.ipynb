# **üü° Intermediate-Level Questions**:

- Write a script to scrape quotes and authors from the first 3 pages.
- How can you extract all the tags associated with each quote?
- The site uses pagination. How would you detect and follow the ‚Äúnext‚Äù page link dynamically?
- How do you scrape the author‚Äôs biography link from each quote card?
- Create a dictionary where the quote text is the key and the author name is the value.


```python
# Write a script to scrape quotes and authors from the first 3 pages.

quotes = []
authors = []

for pno in range(1,4):
    PageURL = f"https://quotes.toscrape.com/page/{pno}/"
    response = requests.get(PageURL)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        for quote in soup.find_all('span', class_="text"):
            quotes.append(quote.text.strip())
        for author in soup.find_all('small', class_="author"):
            authors.append(author.text.strip())
    else:
        print(f"Error fetching page {pno}: {response.status_code}")

print(quotes)
print(authors)            
```

```python
# How can you extract all the tags associated with each quote?

for TAGs in soup.find_all('a',class_='tag'):
    print(TAGs.text.strip())
```

```python
# Pagination of the NEXT button

url = "https://quotes.toscrape.com/"
page = 0

while url:
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    page+=1
    
    for author in soup.find_all('small', class_="author"):
        print(author.text.strip())
    
    next_button = soup.find("li", class_="next")
    
    if next_button:
        next_link = next_button.find("a")["href"]
        url = "https://quotes.toscrape.com/" + next_link

    else:
        url = None
```

```python
# Create a dictionary where the quote text is the key and the author name is the value of 6 pages

dictionary = {}

for pageNo in range(1,6):
    
    PageURL = f"https://quotes.toscrape.com/page/{pageNo}/"
    response = requests.get(PageURL)
    soup = BeautifulSoup(response.text, "html.parser")
    
    if response.status_code == 200:
    
        for qp in soup.find_all('div',class_='quote'):
            key = qp.find('span',class_='text').text
            value = qp.find('small',class_='author').text
        
            dictionary[key] = value
 
print(dictionary)   
```

# üî¥ **Advanced-Level Questions**:

- Write a scraper that collects all quotes across all pages and saves them into a CSV file.


```python
# Save in CSV file

import requests
from bs4 import BeautifulSoup
import csv

base_url = "https://quotes.toscrape.com"
url = "/page/1/"
quotes_data = []

while url:
    response = requests.get(base_url + url)
    soup = BeautifulSoup(response.text, "html.parser")

    for quote in soup.find_all("div", class_="quote"):
        text = quote.find("span", class_="text").get_text(strip=True)
        author = quote.find("small", class_="author").get_text(strip=True)
        tags = [tag.get_text(strip=True) for tag in quote.find_all("a", class_="tag")]

        quotes_data.append({
            "quote": text,
            "author": author,
            "tags": ", ".join(tags)
        })

    next_button = soup.find("li", class_="next")
    url = next_button.find("a")["href"] if next_button else None

with open("quotes.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=["quote", "author", "tags"])
    writer.writeheader()
    writer.writerows(quotes_data)

print("Scraping complete! Quotes saved to quotes.csv")
```
